### VARIABLES AND PREDICTORS

These variables are direct factors that may increase the risks of heart disease in a person.

- Resting blood pressure
    
    Elevated heart rate is associated with elevated blood pressure, increased risk of hypertension, which therefore results in an increased risk of heart disease. 

- Cholesterol levels

    With high cholesterol, a person can develop fatty deposits in their blood vessels. Eventually, these deposits grow, making it difficult for enough blood to flow through your arteries.
    
- Age

    Aging can cause changes in the heart and blood vessels that may increase a person's risk of developing heart disease. 
    
- ST Levels

   ST Levels are a classifier for heart diseases. They help identify if a person is at risk of heart disease by reading their heart rate on an ECG (electrocardiogram) monitor.

### STEPS

After cleaning and tidying the data, we split it into training and testing sets. Knn Classification is used because it is the best algorithm to classify new data points that based on similar measures of the existing  data points.

#### STEP 1:

    Creating a model with the training set using KNN classification. Before applying k-nearest neighbour to a classification task, scaling the data is necessary because it will ensure that all data observations will be on a comparable scale and contribute equal shares to the calculation of the distance between points.

#### STEP 2:

    Identify the most optimal value of k for our data set for k-nn classification using cross validation. Cross-validation is usually used for improving model prediction when we don't have enough data to apply other more efficient methods like the 2-way split (train and test). 
#### STEP 3:

    Create final model with the specific value of k. A final machine learning model is a model that can be used to make predictions on new data.
    
#### STEP 4:

    Test the model by its ability to be able to predict values correctly.
